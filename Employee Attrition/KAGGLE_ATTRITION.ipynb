## Feature Engineering


```python
x=df_new.drop(columns='Attrition')
```

```python
y=df_new['Attrition']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
x_scale=scale.fit_transform(x)
```

```python
conda install -c conda-forge imbalanced-learn
```

```python
from imblearn.under_sampling import NearMiss
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
train_x,test_x,train_y,test_y= train_test_split(x ,y ,test_size=0.3,random_state=101)
```

```python
def model_sel(model,x, y):
    train_x,test_x,train_y,test_y= train_test_split(x ,y ,test_size=0.3,random_state=101)
    model.fit(train_x,train_y)
    train_pred=model.predict(train_x)
    test_pred=model.predict(test_x)
    print("train_accuracy: ", accuracy_score(train_y,train_pred))
    print("test_accuracy: ", accuracy_score(test_y,test_pred))
```

```python
model_sel(log,x,y)
```

```python
model_sel(log,x_scale,y)
```

```python
knn=KNeighborsClassifier()
model_sel(knn,x,y)
```

```python
model_sel(knn,x_scale,y)
```

```python
nb=GaussianNB()
model_sel(nb,x,y)
```

```python
model_sel(nb,x_scale,y)
```

## to find K

```python
error_rate=[]

for i in range(1,40):
    knn=KNeighborsClassifier(i)
    knn.fit(train_x,train_y)
    pred_i=knn.predict(test_x)
    error_rate.append(np.mean(pred_i != test_y))
```

```python
error_rate
```

```python
plt.figure(figsize=(10,4))
plt.plot(range(1,40),error_rate,linestyle="dashed",marker="o",markerfacecolor="red")
```

```python
knn=KNeighborsClassifier(38)
```

```python
model_sel(knn,x_scale,y)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(min_samples_split=100)
```

```python
dt.fit(train_x,train_y)
```

```python
model_sel(dt,x,y)
```

```python


for i in range(5,12,1):
    dt=DecisionTreeClassifier(max_depth=i)
    dt.fit(train_x,train_y)
    y_pred_test=dt.predict(test_x)
    y_pred_train=dt.predict(train_x)
    print("accuracy score for train ---",accuracy_score(train_y,y_pred_train),"---- when min_sam=---",i)
    print("accuracy score for test ---",accuracy_score(test_y,y_pred_test),"---- when min_sam=---",i)
    print("\n")
```
